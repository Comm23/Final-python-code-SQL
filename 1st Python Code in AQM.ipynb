```python
# Multiple linear regressions in Python using the statsmodels library
import pandas as pd
import numpy as np
import random
import itertools
import statsmodels.api as sm 
#Import asset price and liquidity charts
df2 = pd.read_excel(r'C:\Users\User\Downloads\Essai python.xlsx')
vol = pd.read_excel(r'C:\Users\User\Downloads\Essai python volume.xlsx')
df2
# List of dependent variables
dependent_var_col =select_df2.columns

# Dictionary to store regression results for each dependent variable
results_dict = {}

# Loop for each dependent variable
for dependent_var in dependent_var_col:
    # Create a table to record the results of the regressions for this dependent variable
    results_table = pd.DataFrame(columns=['Variables_explicatives', 'Coefficients', 'Erreurs_standard', 'P-value', 'Autocorr_test', 'Hetero_test', 'Multicollinearity_Fstat', 'Multicollinearity_pvalue'])
    
    # Loop for each combination of explanatory variables
    for k in range(1, len(explan_var_col) + 1):
        combinations = itertools.combinations(explan_var_col, k)

        # Loop for each combination of explanatory variables
        for combination in combinations:
            explanatory_subset = list(combination)

            # Run OLS regression
            X = sm.add_constant(merged_data[explanatory_subset])
            y = merged_data[dependent_var]
            model = sm.OLS(y, X)
            results = model.fit()

            # Calculate autocorrelation test (e.g. Durbin-Watson test)
            resid = results.resid
            autocorr_test = sm.stats.stattools.durbin_watson(resid)

            # Compute heteroscedasticity test (e.g. Breusch-Pagan test)
            het_test = sm.stats.diagnostic.het_breuschpagan(resid, X)

            # Calculate multicollinearity test (e.g. F statistic and p-value)
            multicollinearity_Fstat = results.fvalue
            multicollinearity_pvalue = results.f_pvalue

            # Save results to array
            results_table = results_table.append({
                'Variables_explicatives': explanatory_subset,
                'Coefficients': results.params,
                'Erreurs_standard': results.bse,
                'P-value': results.pvalues,
                'Autocorr_test': autocorr_test,
                'Hetero_test': het_test,
                'Multicollinearity_Fstat': multicollinearity_Fstat,
                'Multicollinearity_pvalue': multicollinearity_pvalue
            }, ignore_index=True)
    
   # Add results array to dictionary with dependent variable as key
    results_dict[dependent_var] = results_table

# Display results tables for each dependent variable
for dependent_var, results_table in results_dict.items():
    print(f"Résultats pour la variable dépendante : {dependent_var}")
    print(results_table)
    print() 
```
